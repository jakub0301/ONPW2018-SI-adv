```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf = KNeighborsClassifier(3)
titanic_data = df2.loc[:, df2.columns != 'survived']
titanic_target = df2.survived
```

```python
cross_val_score(clf, titanic_data, titanic_target, cv=5).mean()
```

```python
who_dummies= pd.get_dummies(df2.who, prefix='who')
```

```python
# jeżeli chcemy połączyć ramki danych po unikalnych wartościach w jakichś kolumnach to -> googluj `join in pandas`
df4 = pd.merge(df2, who_dummies, left_index=True, right_index=True).drop(columns='who')
```

```python
deck_dummies = pd.get_dummies(df2.deck, prefix='deck')
town_dummies = pd.get_dummies(df2.embark_town, prefix='town')
class_dummies= pd.get_dummies(df2.pclass, prefix='class')
```

```python
df5 = pd.merge(df4, deck_dummies, left_index=True, right_index=True).drop(columns=['deck'])
df5 = pd.merge(df5, town_dummies, left_index=True, right_index=True).drop(columns=['embark_town'])
df5 = pd.merge(df5, class_dummies,left_index=True, right_index=True).drop(columns=['pclass'])
```

```python
clf = KNeighborsClassifier(3)
titanic_data = df5.loc[:, df5.columns != 'survived']
titanic_target = df5.survived
scores = cross_val_score(clf, titanic_data, titanic_target, cv=5)
scores.mean(), scores.std()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
titanic_data = df5.loc[:, df5.columns != 'survived']
titanic_target = df5.survived
scores = cross_val_score(clf, titanic_data, titanic_target, cv=5)
scores.mean(), scores.std()
```

```python
clf = DecisionTreeClassifier()
titanic_data = df_basic.loc[:, df_basic.columns != 'survived']
titanic_target = df_basic.survived
scores = cross_val_score(clf, titanic_data, titanic_target, cv=5)
scores.mean(), scores.std()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
clf = GradientBoostingClassifier()
titanic_data = df_basic.loc[:, df_basic.columns != 'survived']
titanic_target = df_basic.survived
scores = cross_val_score(clf, titanic_data, titanic_target, cv=5)
scores.mean(), scores.std()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
clf = GradientBoostingClassifier()
X = df_basic.loc[:, df_basic.columns != 'survived']
y = df_basic.survived
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
confusion_matrix(y_test, y_pred)
```
